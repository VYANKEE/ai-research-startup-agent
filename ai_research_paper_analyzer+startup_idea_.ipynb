```python
import requests
import os

API_KEY = os.environ.get("NVIDIA_API_KEY")

def summarize_paper(text):

    prompt = f"""
You are an expert research analyst.

Summarize this research paper clearly.

Provide:
- Title (guess if not given)
- Key Summary
- Core Innovation

Text:
{text[:3000]}
"""

    url = "https://integrate.api.nvidia.com/v1/chat/completions"

    headers = {
        "Authorization": f"Bearer {API_KEY}",
        "Content-Type": "application/json"
    }

    payload = {
        "model": "meta/llama3-70b-instruct",
        "messages": [
            {"role": "user", "content": prompt}
        ],
        "temperature": 0.5,
        "max_tokens": 600
    }

    response = requests.post(url, headers=headers, json=payload)

    result = response.json()

    summary = result["choices"][0]["message"]["content"]

    return summary
```

```python
def find_problems(text):

    prompt = f"""
You are a research problem analyst.

Identify real-world problems this research solves.

Provide:
- Problem 1
- Problem 2
- Problem 3

Text:
{text[:3000]}
"""

    url = "https://integrate.api.nvidia.com/v1/chat/completions"

    headers = {
        "Authorization": f"Bearer {API_KEY}",
        "Content-Type": "application/json"
    }

    payload = {
        "model": "meta/llama3-70b-instruct",
        "messages": [
            {"role": "user", "content": prompt}
        ],
        "temperature": 0.6,
        "max_tokens": 500
    }

    response = requests.post(url, headers=headers, json=payload)

    result = response.json()

    problems = result["choices"][0]["message"]["content"]

    return problems
```

```python
def generate_startup(text):

    prompt = f"""
You are a startup innovation expert.

Based on this research, generate startup ideas.

Provide:
- Startup Name
- Startup Idea
- Target Users
- Value Proposition

Text:
{text[:3000]}
"""

    url = "https://integrate.api.nvidia.com/v1/chat/completions"

    headers = {
        "Authorization": f"Bearer {API_KEY}",
        "Content-Type": "application/json"
    }

    payload = {
        "model": "meta/llama3-70b-instruct",
        "messages": [
            {"role": "user", "content": prompt}
        ],
        "temperature": 0.7,
        "max_tokens": 600
    }

    response = requests.post(url, headers=headers, json=payload)

    result = response.json()

    startup = result["choices"][0]["message"]["content"]

    return startup
```

```python
def generate_business_model(text):

    prompt = f"""
You are a startup business strategist.

Generate a business model.

Provide:
- Revenue Model
- Pricing Strategy
- Market Opportunity
- Scalability Plan

Text:
{text[:3000]}
"""

    url = "https://integrate.api.nvidia.com/v1/chat/completions"

    headers = {
        "Authorization": f"Bearer {API_KEY}",
        "Content-Type": "application/json"
    }

    payload = {
        "model": "meta/llama3-70b-instruct",
        "messages": [
            {"role": "user", "content": prompt}
        ],
        "temperature": 0.6,
        "max_tokens": 600
    }

    response = requests.post(url, headers=headers, json=payload)

    result = response.json()

    business = result["choices"][0]["message"]["content"]

    return business
```

```python
def full_pipeline(file):

    reader = PdfReader(file)

    text = ""

    for page in reader.pages:

        extracted = page.extract_text()

        if extracted:
            text += extracted

    preview = text[:2000]

    summary = summarize_paper(text)

    problems = find_problems(text)

    startup = generate_startup(text)

    business = generate_business_model(text)

    return preview, summary, problems, startup, business
```

```python
with gr.Blocks(theme=gr.themes.Soft()) as demo:

    gr.Markdown("""
    # 🚀 AI Research-to-Startup Generator
    Transform Research Papers into Startup Ideas
    """)

    file_input = gr.File(
        label="📄 Upload Research Paper",
        file_types=[".pdf"]
    )

    run_btn = gr.Button(
        "🚀 Generate Startup Insights",
        variant="primary"
    )

    with gr.Tab("📄 Text Preview"):
        preview_box = gr.Textbox(lines=20)

    with gr.Tab("🧠 Summary"):
        summary_box = gr.Textbox(lines=20)

    with gr.Tab("🎯 Problems"):
        problem_box = gr.Textbox(lines=20)

    with gr.Tab("💡 Startup Idea"):
        startup_box = gr.Textbox(lines=20)

    with gr.Tab("📊 Business Model"):
        business_box = gr.Textbox(lines=20)

    run_btn.click(
        full_pipeline,
        inputs=file_input,
        outputs=[
            preview_box,
            summary_box,
            problem_box,
            startup_box,
            business_box
        ]
    )

demo.launch()
```
